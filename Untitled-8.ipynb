```python
import sklearn
from sklearn import datasets
import pandas as pd
```

```python
boston=datasets.load_boston();
```

```python
df=pd.DataFrame(data=boston.data,columns=boston.feature_names)
df['target']=boston.target
```

```python
df
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
```

```python
y=df.target
X=df.drop('target',axis=1)
X_train,X_test,y_train,y_test=train_test_split(X,y)

```

```python
model = keras.Sequential(
    [
        layers.Dense(500, activation="sigmoid"),
        layers.Dense(800, activation="sigmoid"),
        layers.Dense(4000, activation="sigmoid"),        
        layers.Dense(2000, activation="sigmoid"),
        layers.Dense(1, name="layer5"),
    ]
)
model.compile(optimizer='nadam', loss='mean_absolute_error')
model.fit(X_train, y_train, epochs=100)
```

```python
model.predict((X_test))
```
